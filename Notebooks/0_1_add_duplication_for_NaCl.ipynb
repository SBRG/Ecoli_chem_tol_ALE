```python
import pandas as pd
```

```python
df_NaCl = pd.read_csv("../Data/Mutation_data/Current_AleDB_mut_call/TOL NaCl_mut.csv")
```

```python
end_point_samples=['TOL NaCl A1 F82 I1 R1',
                   'TOL NaCl A2 F82 I1 R1', 
                   'TOL NaCl A3 F82 I1 R1',  
                   'TOL NaCl A4 F82 I1 R1', 
                   'TOL NaCl A5 F82 I1 R1', 
                   'TOL NaCl A6 F85 I1 R1']
```

```python
#starting mutation removal
```

```python
df_NaCl=df_NaCl.drop(index=df_NaCl[df_NaCl['TOL NaCl A1 F0 I1 R1']==1].index)
```

```python
df_NaCl_end=df_NaCl[end_point_samples].fillna(0).astype('int64')
df_NaCl_end=df_NaCl_end.drop(index=df_NaCl_end[df_NaCl_end.sum(axis=1)==0].index)
```

```python
for i in df_NaCl_end.index:
    df_NaCl_end.loc[i,'Position']=df_NaCl.loc[i,'Position']
    df_NaCl_end.loc[i,'Mutation Type']=df_NaCl.loc[i,'Mutation Type']
    df_NaCl_end.loc[i,'Sequence Change']=df_NaCl.loc[i,'Sequence Change']
    df_NaCl_end.loc[i,'Gene']=df_NaCl.loc[i,'Gene']
    df_NaCl_end.loc[i,'Function']=df_NaCl.loc[i,'Function']
    df_NaCl_end.loc[i,'Product']=df_NaCl.loc[i,'Product']
    df_NaCl_end.loc[i,'GO Process']=df_NaCl.loc[i,'GO Process']
    df_NaCl_end.loc[i,'GO Component']=df_NaCl.loc[i,'GO Component']
    df_NaCl_end.loc[i,'Protein change']=df_NaCl.loc[i,'Details']
```

```python
new_col = ['Position', 'Mutation Type', 'Sequence Change', 'Gene',
       'Function', 'Product', 'GO Process', 'GO Component', 'Protein change','TOL NaCl A1 F82 I1 R1',  'TOL NaCl A2 F82 I1 R1', 'TOL NaCl A3 F82 I1 R1',  'TOL NaCl A4 F82 I1 R1', 'TOL NaCl A5 F82 I1 R1', 'TOL NaCl A6 F85 I1 R1']
```

```python
df_NaCl_end=df_NaCl_end[new_col].reset_index(drop=True)
```

```python
dups_dir = "../../ALE_3strain_salt_growth/AMP_dup_run/3Strain_Salt_TALE/MG1655_NaCl_TALE/breseq/dups/"
```

```python
import os

dir_content = os.listdir(dups_dir)
sample_list = []

for sub in dir_content:
    d = os.path.join(dups_dir,sub)
    if os.path.isdir(d)==True:
        sample_list.append(sub)
    
```

```python
sample_list
```

```python
#Convert breseq directory names to column names used in Variant call files

col_names=pd.DataFrame(index=sample_list)
col_names['col_names']=col_names.index
col_names=col_names.drop(index='1-0-1-1') #ignore first 

col_names[['A','F','I','R']]=col_names['col_names'].str.split("-", expand=True)
col_names=col_names.sort_values(by="A")

for i in col_names.index:
    col_names.loc[i,'col_names']="TOL NaCl A"+col_names.loc[i,'A']+" F"+col_names.loc[i,'F']+" I"+col_names.loc[i,'I']+" R1"
```

```python
col_names
```

```python
for file in col_names.index:

    duplication_df_temp = pd.read_csv(dups_dir+file+"/"+file+"_genes.csv")
    
    if duplication_df_temp.index.size==0:
        continue
    
    k=0    
    for dup_count in duplication_df_temp.index:
        
        temp_row = pd.DataFrame()
        temp_row.loc[k,'Position'] = str(duplication_df_temp.loc[dup_count,'Start'])
        temp_row.loc[k,'Mutation Type'] = 'DUP'
        temp_row.loc[k,'Sequence Change'] = format (duplication_df_temp.loc[dup_count,'Width'], ',d')+" bp x"+str(duplication_df_temp.loc[dup_count,'Multiplicity'])
        temp_row.loc[k,'Gene']=duplication_df_temp.loc[dup_count,'Gene List'].replace("['","").replace("']","").replace("'","")
        temp_row.loc[k,'Protein change']="Duplication"
        temp_row.loc[k,col_names.loc[file,'col_names']]=1
        temp_row=temp_row.fillna("")
        df_NaCl_end=df_NaCl_end.append(temp_row)
            
```

```python
df_NaCl_end['Position'].str.replace(",","")
```

```python
df_NaCl_end['Position']=df_NaCl_end['Position'].str.replace(",","")
df_NaCl_end=df_NaCl_end.sort_values(by="Position").reset_index(drop=True)
```

```python
for i in df_NaCl_end.index:
    df_NaCl_end.loc[i,'Position']=format(int(df_NaCl_end.loc[i,'Position']),",")
```
